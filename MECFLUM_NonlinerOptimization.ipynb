#### Identification of BEP using Newton-Raphson Method with Multiple Equations & Unknowns


```python
m = GEKKO()
p = m.Var(value=800)    
w = m.Var(value=5) 
w1 = m.Var(value=3)
w2 = m.Var(value=1.5)
m.Equations([p-7.2*w**2-392.4==0, p-810+3.75*w1**2+25*w1==0, p-900+30*w2**2+65*w2==0, w-w1-w2==0]) # equations
m.solve(disp=False)     # solve
print([p.value[0],w.value[0],w1.value[0],w2.value[0]]) # print solution
```

### MVNR Implementation on Hardy-Cross 


```python
m = GEKKO()
m.solver_options = ['minlp_gap_tol 1.0e-5',\
                    'minlp_maximum_iterations 10000']
m.options.solver = 1 #uses APOPT (Advanced Process OPTimizer) used for Nonlinear Programming. Other solvers do not accept 7 equations for 6 unknowns
x1 = m.Var(value=0)
x2 = m.Var(value=0) 
x3 = m.Var(value=0) 
x4 = m.Var(value=0) 
x5 = m.Var(value=0) 
x6 = m.Var(value=0) 
k1, k2, k3, k4, k5, k6 = [1,3,4,4,5,2]
m.Equations([x1+x2==100, x1==x3+x4, x4==x6+25, x2+x3==x5+25, x5+x6==50, k1*x1**2+k3*x3**2-k2*x2**2==0, k4*x4**2+k6*x6**2-k5*x5**2-k3*x3**2==0])
m.solve(disp=False)     # solve
print([x1.value[0],x2.value[0],x3.value[0],x4.value[0],x5.value[0],x6.value[0]]) # print solution
```
